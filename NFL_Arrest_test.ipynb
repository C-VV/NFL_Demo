```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
```

```python
base_url = "http://nflarrest.com/api/v1/team/timeline"
header= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:70.0) Gecko/20100101 Firefox/70.0'}
```

```python
city_codes = ["MIN","DEN","CIN","JAC","TB","TEN","CLE","IND","KC","CHI","MIA","SEA","SF","BAL","LAC","ARI","NO","GB","PIT","NYJ","OAK",
"WAS","BUF","DAL","NE","CAR","NYG","ATL","PHI","LA","DET","HOU"]
name = []
team_dict = []

team = 'dal'
query_url = (f'{base_url}/{team}')
response_json = requests.get(query_url,headers=header).json()
response_json

for team in city_codes:
    print(f"processing team: {team} ")
    query_url = (f'{base_url}/{team}')
    arrests_json = requests.get(query_url,headers=header).json()
    for arrest in arrests_json:
        print(f"Team: {team}, Month: {arrest['Month']}, Year: {arrest['Year']}, Count: {arrest['arrest_count']}")
        new_row={}
        new_row['Team'] = team
        new_row['Month'] = arrest['Month']
        new_row['Year'] = arrest['Year']
        new_row['arrest_count'] = arrest['arrest_count']
        team_dict.append(new_row)
```

```python
#team_dict[0][0]
#nested for loop to associate team with the records
team_dict
```

```python
team_arrests = pd.DataFrame(team_dict)
team_arrests
```

```python
#Create dateframes to exclude 2000,2001 and 2019 seasons
ta2000_df = team_arrests[team_arrests.Year != "2000"]
ta2001_df = ta2000_df[ta2000_df.Year != "2001"]
ta2017_df = ta2001_df[ta2001_df.Year != "2017"]
ta2018_df = ta2017_df[ta2017_df.Year != "2018"]
ta_df = ta2018_df[ta2018_df.Year != "2019"]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(ta_df)
```
